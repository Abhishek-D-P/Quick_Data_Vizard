```python
import numpy as np
from LinearRegression import LinearRegression
```

```python
rows=1000
columns=2
```

```python
X=np.random.rand(rows,columns)
# Y_train=np.random.rand(rows,1)+np.dot(X_train,np.array([2,1,1]).reshape(3,1))
y=X[:,0]*20+X[:,1]*15+np.random.rand()+10
```

```python
import pandas as pd

data=pd.DataFrame({"f1":X[:,0],'f2':X[:,1],'y':y})
data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],y)
```

```python
def train_test_split(X,y,ratio=0.2):
    train_index=np.random.choice([i for i in range(len(X))],size=int(ratio*len(X)))
    X_train=X[train_index]
    y_train=y[train_index]

    X_test=X[~train_index]
    y_test=y[~train_index]

    return X_train,y_train,X_test,y_test

```

```python
X_train,y_train,X_test,y_test=train_test_split(X,y,0.2)
```

```python
model=LinearRegression()
weights,bias,loss=model.fit(X_train,y_train,X_test,y_test)
```

```python
plt.plot(loss)
```

```python
print(weights,bias)
```

```python
print(model.score(X_test,y_test))
print(model.score(X_train,y_train))
```
